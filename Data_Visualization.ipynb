```python

import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
df = pd.read_excel("Online-Store-Orders.xlsx")
print(df.head())
print(df.info())
print(df.isnull().sum())
df.hist(figsize=(10,8))
plt.show()
df.groupby("Product")["TotalPrice"].sum().plot(kind="bar", color="skyblue")
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Price")
plt.show()
df["PaymentMethod"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Payment Method Distribution")
plt.ylabel("")
plt.show()
df.groupby("Product")["Quantity"].sum().plot(kind="line", marker="o")
plt.title("Quantity Sold by Product")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.grid(True)
plt.show()
```
